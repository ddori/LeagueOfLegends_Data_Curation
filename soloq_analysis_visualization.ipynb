```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

csv_path = 'SoloQ/data/soloq_full_1520.csv'
df = pd.read_csv(csv_path, low_memory=False)
df['tier'] = df['tier'].astype(str).str.strip().str.upper()
tier_order = ["IRON","BRONZE","SILVER","GOLD","PLATINUM","EMERALD","DIAMOND","MASTER","GRANDMASTER","CHALLENGER"]
present_tiers = [t for t in tier_order if t in set(df['tier'])] or sorted(df['tier'].unique().tolist())
df['tier'] = pd.Categorical(df['tier'], categories=present_tiers, ordered=True)
safe_deaths = df['deaths'].replace(0, np.nan)
df['kda_robust'] = (df['kills'] + df['assists']) / safe_deaths
df['kda_robust'] = df['kda_robust'].replace([np.inf, -np.inf], np.nan).fillna(df['kills'] + df['assists'])
df['gameMinutes'] = np.where(df.columns.str.contains('gameDuration').any(), df['gameDuration']/60.0, np.where(df.columns.str.contains('timePlayed').any(), df['timePlayed']/60.0, np.nan))
agg = df.groupby('tier').agg(rows=('matchId','size')).reset_index()
plt.figure(figsize=(9,5))
plt.bar(agg['tier'].astype(str), agg['rows'])
plt.title("Rows per Tier (participant-rows)")
plt.xlabel("Tier")
plt.ylabel("Rows")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

```python

agg = df.groupby('tier').agg(kda_median=('kda_robust','median')).reset_index()
plt.figure(figsize=(9,5))
plt.plot(agg['tier'].astype(str), agg['kda_median'], marker='o')
plt.title("Median Robust KDA by Tier")
plt.xlabel("Tier")
plt.ylabel("Median KDA")
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
```

```python
agg = df.groupby('tier').agg(gpm_mean=('gpm','mean')).reset_index()
plt.figure(figsize=(9,5))
plt.plot(agg['tier'].astype(str), agg['gpm_mean'], marker='o')
plt.title("Gold per Minute (mean) by Tier")
plt.xlabel("Tier")
plt.ylabel("GPM")
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
```

```python
agg = df.groupby('tier').agg(dpm_mean=('dpm','mean')).reset_index()
plt.figure(figsize=(9,5))
plt.plot(agg['tier'].astype(str), agg['dpm_mean'], marker='o')
plt.title("Damage per Minute (mean) by Tier")
plt.xlabel("Tier")
plt.ylabel("DPM")
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
```

```python

agg = df.groupby('tier').agg(cspm_mean=('cspm','mean')).reset_index()
plt.figure(figsize=(9,5))
plt.plot(agg['tier'].astype(str), agg['cspm_mean'], marker='o')
plt.title("CS per Minute (mean) by Tier")
plt.xlabel("Tier")
plt.ylabel("CSPM")
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
```

```python

agg = df.groupby('tier').agg(game_min_median=('gameMinutes','median')).reset_index()
plt.figure(figsize=(9,5))
plt.plot(agg['tier'].astype(str), agg['game_min_median'], marker='o')
plt.title("Game Length (median, minutes) by Tier")
plt.xlabel("Tier")
plt.ylabel("Minutes")
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
```

```python

pos_col = 'teamPosition' if 'teamPosition' in df.columns else ('individualPosition' if 'individualPosition' in df.columns else None)
pos_map = {'TOP':'TOP','JUNGLE':'JUNGLE','MIDDLE':'MID','MID':'MID','BOTTOM':'ADC','ADC':'ADC','CARRY':'ADC','UTILITY':'SUPPORT','SUPPORT':'SUPPORT','NONE':'NONE','FILL':'FILL'}
pos_series = df[pos_col].astype(str).str.upper().map(pos_map).fillna('OTHER')
tmp = df.copy()
tmp['_pos_norm'] = pos_series
pos_pivot = tmp.groupby(['tier','_pos_norm'])['puuid'].nunique().reset_index(name='players')
pos_pivot['players_total'] = pos_pivot.groupby('tier')['players'].transform('sum')
pos_pivot['share'] = pos_pivot['players'] / pos_pivot['players_total']
pos_order = ['TOP','JUNGLE','MID','ADC','SUPPORT','FILL','NONE','OTHER']
table = pos_pivot.pivot(index='tier', columns='_pos_norm', values='share').reindex(columns=pos_order, fill_value=0.0).reindex(present_tiers)

plt.figure(figsize=(10,6))
bottom = np.zeros(len(table))
x = np.arange(len(table.index))
for role in table.columns:
    plt.bar(x, table[role].values, bottom=bottom, label=role)
    bottom += table[role].values
plt.title("Role Composition by Tier (share of unique players)")
plt.xlabel("Tier")
plt.ylabel("Share")
plt.xticks(x, table.index.astype(str), rotation=45, ha='right')
plt.legend(ncol=4, fontsize=8)
plt.tight_layout()
```

```python
df = df.copy()
df['teamId'] = df['team_teamId']
team_cols = ['team_obj_dragon_kills','team_obj_baron_kills','team_obj_tower_kills','team_obj_riftHerald_kills']
av = [c for c in team_cols if c in df.columns]
team_df = df[['matchId','teamId','tier'] + av].drop_duplicates(subset=['matchId','teamId'])
team_agg = team_df.groupby('tier')[av].mean().reset_index()
plt.figure(figsize=(10,6))
for col in av:
    plt.plot(team_agg['tier'].astype(str), team_agg[col], marker='o', label=col)
plt.title("Team Objectives per Game by Tier (mean)")
plt.xlabel("Tier")
plt.ylabel("Per-game mean (team level)")
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend()
plt.tight_layout()
```

```python
div = df.groupby('tier')['championName'].nunique().reset_index(name='unique_champions')
plt.figure(figsize=(9,5))
plt.bar(div['tier'].astype(str), div['unique_champions'])
plt.title("Champion Diversity by Tier (unique champions played)")
plt.xlabel("Tier")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

```python
tmp = df.copy()
tmp['teamDamagePct'] = tmp['ch_teamDamagePercentage'] * 100.0
agg = tmp.groupby('tier').agg(teamDmgPct_median=('teamDamagePct','median'), kda_median=('kda_robust','median')).reset_index()
plt.figure(figsize=(7,6))
plt.scatter(agg['teamDmgPct_median'], agg['kda_median'])
for i, row in agg.iterrows():
    plt.annotate(str(row['tier']), (row['teamDmgPct_median'], row['kda_median']))
plt.title("Tier median: Damage Share vs Robust KDA")
plt.xlabel("Team Damage Share (%) - median")
plt.ylabel("Robust KDA - median")
plt.tight_layout()
```
